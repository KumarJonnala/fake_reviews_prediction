```python
import requests
```

```python
model = "deepseek-r1:1.5b"# specify the model name
```

```python
# Function to query the local LLM server

def query_llm(prompt: str, model: str = model) -> str: # arguments promt and model
    # Send a request to the local LLM server, stream set to Falsefor single responce
    response = requests.post(
        "http://localhost:11434/api/generate",  
        json={"model": model, "prompt": prompt, "stream": False}
    )
    return response.json()["response"]
```

```python
prompt = """Classify the sentiment of below statement into Original reviews, Computer-generated.
'Love this!  Well made, sturdy, and very comfortable.  I love it!Very pretty' """
```

```python
print(query_llm(prompt))
```
